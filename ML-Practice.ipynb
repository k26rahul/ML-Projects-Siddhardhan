## numpy.linspace

`numpy.linspace(start, stop, num=50, endpoint=True, retstep=False)`

- Returns num evenly spaced samples, calculated over the interval [start, stop].
- The endpoint of the interval can optionally be excluded.

Optional parameters:
- `endpoint: bool, optional`
  - If True, stop is the last sample. Otherwise, it is not included. Default is True.
- `retstep: bool, optional`
  - If True, return (samples, step), where step is the spacing between samples.

````python
np.linspace(0, 2, 20)
````

````python
np.linspace(0, 2, 20, endpoint=False, retstep=True)
````

## numpy.arange

`numpy.arange(start, stop, step)`

- Return evenly spaced values within a given interval.
- Values are generated within the half-open interval `[start, stop)`.
- For integer arguments the function is roughly equivalent to the Python built-in range, but returns an ndarray rather than a range instance.

Understanding the official function signature:
```
numpy.arange(
    [start, ]   # Optional parameter, default is 0 if not provided
    stop,       # Required parameter, the end value of the sequence
    [step, ]    # Optional parameter, default is 1 if not provided
    dtype=None, # Optional parameter, data type of the output array, default is inferred
    *,          # Separator indicating that the following parameters must be provided using keyword syntax
    like=None   # Optional parameter, an existing array-like object for properties of the output array
)
```


````python
np.arange(16)
np.arange(5, 16)
np.arange(5, 16, 3)
````

## numpy.geomspace

`numpy.geomspace(start, stop, num=50, endpoint=True, dtype=None, axis=0)`

- Return numbers spaced evenly on a log scale (a geometric progression).
- Used to generate numbers in a geometric progression.
- This is similar to `logspace`, but with endpoints specified directly. Each output sample is a constant multiple of the previous.

### **Explanation 1**

#### Logarithmic Scale:

A logarithmic scale is a way of representing data in which values increase or decrease exponentially. Instead of representing each value linearly, as you would on a regular (linear) scale, a logarithmic scale represents values in powers of a logarithm. The two most common logarithmic scales are the base-10 logarithmic scale (common logarithm) and the natural logarithmic scale (base e).

On a logarithmic scale, each unit represents a power of the logarithm base. For example, on a base-10 logarithmic scale, each unit represents a power of 10. This means that the distance between 1 and 10 is the same as the distance between 10 and 100, and so on.

#### Geometric Progression:

A geometric progression (or geometric sequence) is a sequence of numbers where each term after the first is found by multiplying the previous term by a fixed, non-zero number called the common ratio. The general form of a geometric progression is:

$ a, ar, ar^2, ar^3, \ldots $

where:
- $a$ is the first term,
- $r$ is the common ratio.

In a geometric progression, each term is the product of the previous term and a constant ratio. This leads to an exponential growth or decay pattern.

#### Connection:

The connection between a logarithmic scale and a geometric progression lies in the fact that when you plot the terms of a geometric progression on a logarithmic scale, the resulting graph appears linear. This is because the logarithmic scale compresses the exponential growth or decay of the geometric progression into a linear representation.

In the context of `numpy.geomspace()`, it generates values that are evenly spaced on a logarithmic scale, creating a geometric progression.

For example, if you have a geometric progression with a common ratio of 10 (like 1, 10, 100, 1000), when plotted on a logarithmic scale, the intervals between the values will be constant, creating a straight line. The `numpy.geomspace()` function helps in generating such sequences efficiently.

### **Explanation 2**

#### Log Scale:

A logarithmic scale is a way of representing data in which the values on the scale are not evenly spaced, but the intervals represent orders of magnitude. In other words, each tick mark on a logarithmic scale represents a multiple of some base raised to a power.

The logarithm of a number is the exponent to which a fixed base (usually 10 or sometimes e, the base of the natural logarithm) must be raised to obtain that number. Mathematically, if $ b^x = y $, then $ \log_b(y) = x $.

For example, on a log scale, each successive tick mark might represent a tenfold increase. This is particularly useful when dealing with data that spans several orders of magnitude, as it compresses the scale and allows for a more informative representation.

#### Geometric Progression:

A geometric progression (or sequence) is a sequence of numbers where each term after the first is found by multiplying the previous term by a fixed, non-zero number called the common ratio. Mathematically, if $ a $ is the first term and $ r $ is the common ratio, then the terms of the sequence are $ a, ar, ar^2, ar^3, \ldots $.

In the context of `numpy.geomspace()`, it's generating a geometric progression where the start value is the first term, the end value is the last term, and the number of samples requested determines the total number of terms in the sequence. The values are spaced evenly on a logarithmic scale, meaning the ratio between successive values is constant.

In summary, when you use `numpy.geomspace()`, you are generating a sequence of numbers that are spaced evenly on a logarithmic scale, forming a geometric progression. This is useful in various scientific and engineering applications where you want to represent data that spans multiple orders of magnitude.

````python
np.geomspace(1, 1000, num=4)
np.geomspace(1, 1000, 4)
````

````python
np.geomspace(1, 1000, 3)
np.geomspace(1, 1000, 3, endpoint=False)
````

````python
np.geomspace(1, 256, 9)
np.geomspace(1, 256, 10)
````

### Graphical illustration:

````python
# Generate a geometric progression manually for comparison
start_manual = 1
stop_manual = 1000
num_terms = 10
common_ratio = (stop_manual / start_manual) ** (1 / (num_terms - 1))

geometric_sequence_manual = [start_manual *
                             common_ratio ** i for i in range(num_terms)]

geometric_sequence_manual
````

````python
# Generate a geometric progression using numpy.geomspace
start_np = 1
stop_np = 1000
num_samples = 10

geometric_sequence_np = np.geomspace(start_np, stop_np, num=num_samples)
geometric_sequence_np
````

````python
# Plot the results
plt.figure(figsize=(12, 5))

# Plot manually generated geometric progression
plt.subplot(1, 2, 1)
plt.plot(geometric_sequence_manual, marker='o', linestyle='-', color='b')
_ = plt.title('Manually Generated Geometric Progression')
_ = plt.xlabel('Term Index')
_ = plt.ylabel('Value')

# Plot numpy.geomspace output
plt.subplot(1, 2, 2)
plt.plot(geometric_sequence_np, marker='o', linestyle='-', color='r')
_ = plt.title('numpy.geomspace() Output')
_ = plt.xlabel('Term Index')
_ = plt.ylabel('Value')

plt.tight_layout()
plt.show()
````

## numpy.logspace

`numpy.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None, axis=0)`

- Return numbers spaced evenly on a log scale.
- In linear space, the sequence starts at `base ** start` (base to the power of start) and ends with `base ** stop`.
- `base ** stop` is the final value of the sequence, unless `endpoint` is False. In that case, `num + 1` values are spaced over the interval in log-space, of which all but the last (a sequence of length num) are returned.

````python
np.logspace(2, 3, num=4)
np.logspace(2, 3, num=3)
np.logspace(2, 3, num=2)
````

````python
np.logspace(1, 10, base=2, num=10)
np.logspace(1, 10, base=2, num=8)
````

````python
np.logspace(1, 10, base=2, num=5)
np.logspace(1, 10, base=2, num=4)
np.logspace(1, 10, base=2, num=3)
np.logspace(1, 10, base=2, num=2)
````

**Exponential Function:**

An exponential function is a mathematical function of the form:

$ f(x) = a \cdot b^x $

Here:

- $ a $ is a constant, representing the initial or starting value of the function.
- $ b $ is a constant, known as the base of the exponential function. It's a positive number and not equal to 1.
- $ x $ is the variable, and it appears in the exponent.

Exponential functions exhibit rapid growth (if $ b > 1 $) or decay (if $ 0 < b < 1 $) as $ x $ increases. The rate of growth or decay is determined by the value of $ b $. If $ b > 1 $, the function grows exponentially; if $ 0 < b < 1 $, the function decays exponentially.

For example:

- $ f(x) = 2^x $ is an exponential growth function with a base of 2.
- $ g(x) = 0.5^x $ is an exponential decay function with a base of 0.5.

Exponential functions are commonly encountered in various scientific, financial, and natural processes where quantities grow or decay at a constant rate proportional to their current value.

In Python, you can use the `numpy.exp()` function to calculate the exponential of an array of values. For example:

```python
import numpy as np

x = np.array([1, 2, 3])
result = np.exp(x)

print(result)
# Output: [ 2.71828183  7.3890561  20.08553692]
```

This calculates the exponential function $ e^x $ for each element in the array `x`.

**Negative exponentiation:**

In mathematics, raising a number to a negative exponent is equivalent to taking the reciprocal of the number raised to the positive exponent. Specifically, if $a$ is a non-zero number and $n$ is a positive integer, then $a^{-n}$ is the same as $\frac{1}{a^n}$.

For example:

- $2^{-3} = \frac{1}{2^3} = \frac{1}{8}$
- $10^{-2} = \frac{1}{10^2} = \frac{1}{100}$
- $e^{-5} = \frac{1}{e^5} \approx \frac{1}{148.413}$ (where $e$ is the base of the natural logarithm)

`np.exp(-t)` represents the exponential decay of a quantity over time, where $t$ is the variable and the decay rate is determined by the magnitude of $t$. As $t$ increases, $e^{-t}$ approaches zero, leading to a decay behavior.

**Damped Oscillation:**

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
```

This function represents a damped oscillation, and you can observe exponential decay in the presence of the term $\exp(-t)$. Let's break down the function and explain each part:

- $ \exp(-t) $: This term represents an exponential decay. As $ t $ increases, the value of $ \exp(-t) $ decreases rapidly, approaching zero. Exponential decay is a common mathematical model for processes where a quantity decreases over time.

- $ \cos(2\pi t) $: This term represents a cosine function with a frequency of $ 2\pi $ cycles per unit of $ t $. The cosine function introduces oscillations, causing the overall behavior to oscillate as it decays.

When you combine these two terms in the function, $ \exp(-t) \cdot \cos(2\pi t) $, you get a damped oscillation where the amplitude of the oscillation decreases over time due to the exponential decay factor. The exponential decay term causes the amplitude of the cosine wave to decrease, and as $ t $ becomes large, the oscillations diminish, and the function tends toward zero.

This type of behavior is often seen in physical systems with damping, where energy is gradually lost to the surroundings. The exponential decay term models the dissipation of energy over time, and the cosine term introduces oscillations.

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
````

## Illustration of a scatter plot

In summary, the code generates a scatter plot with random data, where the x and y coordinates, color, and marker size are all determined by different arrays in the `data` dictionary. The plot is then displayed with specified labels for the x and y axes.

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
```

- `data`: This is a dictionary containing three key-value pairs:
  - `'a'`: An array of integers from 0 to 49 (created using `np.arange(50)`).
  - `'c'`: An array of 50 random integers between 0 (inclusive) and 50 (exclusive) (created using `np.random.randint(0, 50, 50)`).
  - `'d'`: An array of 50 random numbers drawn from a standard normal distribution (mean=0, standard deviation=1) (created using `np.random.randn(50)`).

```python
data['b'] = data['a'] + 10 * np.random.randn(50)
```

- `data['b'] = data['a'] + 10 * np.random.randn(50)`: This line creates a new key `'b'` in the `data` dictionary and assigns it values equal to `'a'` plus 10 times random numbers drawn from a standard normal distribution.

```python
data['d'] = np.abs(data['d']) * 100
```

- `data['d'] = np.abs(data['d']) * 100`: This line updates the values associated with the key `'d'` in the `data` dictionary. It takes the absolute values of the random numbers in `'d'` and multiplies them by 100.

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
```

- `fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')`: This line creates a new figure and axes using `plt.subplots()`. The `figsize` parameter sets the size of the figure, and `layout='constrained'` specifies a constrained layout.

```python
ax.scatter('a', 'b', c='c', s='d', data=data)
```

- `ax.scatter('a', 'b', c='c', s='d', data=data)`: This line creates a scatter plot on the axes `ax`. The x-coordinates are taken from `'a'`, y-coordinates from `'b'`, color is determined by the values in `'c'`, and the size of the markers is determined by the values in `'d'`.

````python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
````
